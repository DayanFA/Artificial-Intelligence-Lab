```python
import tensorflow as tf
from IPython.display import Audio

link = '/kaggle/input/audio-cats-and-dogs/cats_dogs/test/test/dog_barking_73.wav'
arquivo_audio = tf.io.read_file( link )
conteudo, taxa_amostragem = tf.audio.decode_wav(
    arquivo_audio,
    desired_channels = 1
)
Audio(
    data = tf.squeeze( conteudo ),
    rate = taxa_amostragem.numpy(),
)
```

```python
import matplotlib.pyplot as plt
plt.plot(tf.squeeze(conteudo))
```

```python
import librosa

conteudo_tempo, taxa_amostragem = librosa.load( link )
CONTEUDO_FREQ = librosa.stft(conteudo_tempo)
CONTEUDO_FREQ_DB = librosa.amplitude_to_db(abs(CONTEUDO_FREQ))
plt.figure()
librosa.display.specshow(
    CONTEUDO_FREQ_DB,
    sr = taxa_amostragem,
    #y_axis = 'hz'
)
```

```python
import os
pasta_puxar_audio = '/kaggle/input/audio-cats-and-dogs/cats_dogs/test/cats/'
pasta_salvar_fig = '/kaggle/working/test/cat/'

for arquivo in os.listdir(pasta_puxar_audio):
    print(arquivo)
    conteudo_tempo, taxa_amostragem = librosa.load( pasta_puxar_audio + arquivo )
    CONTEUDO_FREQ = librosa.stft(conteudo_tempo)
    CONTEUDO_FREQ_DB = librosa.amplitude_to_db(abs(CONTEUDO_FREQ))
    plt.figure( figsize = (14,5) )
    librosa.display.specshow(
    CONTEUDO_FREQ_DB,
    sr = taxa_amostragem,
    #y_axis = 'hz'
    )
    plt.axis('off')
    plt.savefig(
        pasta_salvar_fig+arquivo.replace('.wav','.jpeg'),
        bbox_inches = 'tight',
        pad_inches = 0,
        )
```
