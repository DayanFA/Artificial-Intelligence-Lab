```python
import tensorflow as tf
from IPython.display import Audio
```

```python
endereco_arquivo ='/kaggle/input/audio-cats-and-dogs/cats_dogs/test/cats/cat_90.wav'
arquivo = tf.io.read_file(endereco_arquivo)
matriz_audio, taxa_amostragem = tf.audio.decode_wav(
    arquivo,
    desired_channels = 1
)
matriz_corrigida = tf.squeeze(matriz_audio)
Audio(
    data = matriz_corrigida,
    rate = taxa_amostragem.numpy()*3,
)
```

```python
import numpy as np
freq_som = 400 #quatidade de picos de meu sinal por seg
taxa_amostragem = 1600 #qtd de  pontos que o altofalante acompanha por seg
tempo = 4 #seg
sinal1 = np.sin(2*np.pi*freq_som*\
              np.linspace(0, tempo, tempo*taxa_amostragem))

freq_som2 = 1000
sinal2 = np.sin(2*np.pi*freq_som2*\
              np.linspace(0, tempo, tempo*taxa_amostragem))
Audio(
    data = (sinal1 + sinal2)/2,
    rate = taxa_amostragem
)
```
