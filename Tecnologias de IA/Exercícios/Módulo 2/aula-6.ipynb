```python
import librosa
import matplotlib.pyplot as plt
import os

pasta_audios = \
    '/kaggle/input/audio-cats-and-dogs/cats_dogs/test/cats/'
pasta_figs = \
    '/kaggle/working/test/cat/'
for arquivo in os.listdir(pasta_audios):
    print(arquivo)
    arquivo_audio, taxa_amo = librosa.load(pasta_audios+arquivo)
    ARQUIVO_AUDIO = librosa.stft(arquivo_audio)
    ARQUIVO_AUDIO_DB = librosa.amplitude_to_db(abs(ARQUIVO_AUDIO))
    plt.figure()
    librosa.display.specshow(ARQUIVO_AUDIO_DB,sr=taxa_amo,y_axis='hz')
    plt.axis('off')
    plt.savefig(pasta_figs+arquivo.replace('.wav','.jpeg'))
    plt.close()
print(os.listdir("test/cat/"))

pasta_audios = \
    '/kaggle/input/audio-cats-and-dogs/cats_dogs/test/test/' #
pasta_figs = \
    '/kaggle/working/test/dog/' #
for arquivo in os.listdir(pasta_audios):
    print(arquivo)
    arquivo_audio, taxa_amo = librosa.load(pasta_audios+arquivo)#
    ARQUIVO_AUDIO = librosa.stft(arquivo_audio)
    ARQUIVO_AUDIO_DB = librosa.amplitude_to_db(abs(ARQUIVO_AUDIO))
    plt.figure()
    librosa.display.specshow(ARQUIVO_AUDIO_DB,sr=taxa_amo,y_axis='hz')
    plt.axis('off')
    plt.savefig(pasta_figs+arquivo.replace('.wav','.jpeg'))
    plt.close()
print(os.listdir("test/dog/")) #

pasta_audios = \
    '/kaggle/input/audio-cats-and-dogs/cats_dogs/train/cat/' #
pasta_figs = \
    '/kaggle/working/train/cat/' #
for arquivo in os.listdir(pasta_audios):
    print(arquivo)
    arquivo_audio, taxa_amo = librosa.load(pasta_audios+arquivo)#
    ARQUIVO_AUDIO = librosa.stft(arquivo_audio)
    ARQUIVO_AUDIO_DB = librosa.amplitude_to_db(abs(ARQUIVO_AUDIO))
    plt.figure()
    librosa.display.specshow(ARQUIVO_AUDIO_DB,sr=taxa_amo,y_axis='hz')
    plt.axis('off')
    plt.savefig(pasta_figs+arquivo.replace('.wav','.jpeg'))
    plt.close()
print(os.listdir("train/cat/")) #

pasta_audios = \
    '/kaggle/input/audio-cats-and-dogs/cats_dogs/train/dog/' #
pasta_figs = \
    '/kaggle/working/train/dog/' #
for arquivo in os.listdir(pasta_audios):
    print(arquivo)
    arquivo_audio, taxa_amo = librosa.load(pasta_audios+arquivo)#
    ARQUIVO_AUDIO = librosa.stft(arquivo_audio)
    ARQUIVO_AUDIO_DB = librosa.amplitude_to_db(abs(ARQUIVO_AUDIO))
    plt.figure()
    librosa.display.specshow(ARQUIVO_AUDIO_DB,sr=taxa_amo,y_axis='hz')
    plt.axis('off')
    plt.savefig(pasta_figs+arquivo.replace('.wav','.jpeg'))
    plt.close()
print(os.listdir("train/dog/")) #

```
